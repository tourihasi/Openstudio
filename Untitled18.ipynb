```python
import csv
from pathlib import Path

import openstudio

OSM_NAME = next(iter(uploaded.keys()))  # アップロードしたファイル名
OSM_PATH = Path(OSM_NAME)
OUT_PATH = Path("zone_space_surface.csv")

def safe_name(obj) -> str:
    try:
        return obj.nameString()
    except Exception:
        return ""

def construction_name(surface: openstudio.model.Surface) -> str:
    opt = surface.construction()
    if opt.is_initialized():
        return safe_name(opt.get())
    return ""

# ---- Load model ----
vt = openstudio.osversion.VersionTranslator()
mopt = vt.loadModel(openstudio.path(str(OSM_PATH)))
if mopt.empty():
    raise RuntimeError("Failed to load OSM (VersionTranslator returned empty model).")

model = mopt.get()

fields = [
    "ThermalZone Name",  # 後工程で必須になりがちなので追加
    "Space Name",
    "Surface Name",
    "Surface Type",
    "Construction",
    "Sun Expose",
    "Wind Expose",
]

rows = 0
with OUT_PATH.open("w", newline="", encoding="utf-8-sig") as f:
    w = csv.DictWriter(f, fieldnames=fields)
    w.writeheader()

    for sp in model.getSpaces():
        sp_name = safe_name(sp)

        tz_name = ""
        tz_opt = sp.thermalZone()
        if tz_opt.is_initialized():
            tz_name = safe_name(tz_opt.get())

        for srf in sp.surfaces():
            w.writerow({
                "ThermalZone Name": tz_name,
                "Space Name": sp_name,
                "Surface Name": safe_name(srf),
                "Surface Type": srf.surfaceType(),
                "Construction": construction_name(srf),
                "Sun Expose": srf.sunExposure(),
                "Wind Expose": srf.windExposure(),
            })
            rows += 1

print("DONE:", rows, "rows")
print("CSV:", OUT_PATH)
```
