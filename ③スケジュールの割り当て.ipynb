```python
#1 Colab用ユーティリティ
!pip -q install openstudio
from google.colab import files
import zipfile, datetime
import openstudio as osd

#2 割当対象名
SPACE_TYPE = "189.1-2009 - Office - BreakRoom - CZ1-3"
CONSTR_SET = "189.1-2009 - CZ1 - Office"
SCHED_SET  = "189.1-2009 - Office - BreakRoom - CZ1-3 Schedule Set 1"

#3 一括割当関数
def assign(m):
    st = m.getSpaceTypeByName(SPACE_TYPE).get()
    cs = m.getDefaultConstructionSetByName(CONSTR_SET).get()
    ss = m.getDefaultScheduleSetByName(SCHED_SET).get()
    for sp in m.getSpaces():
        sp.setSpaceType(st)
        sp.setDefaultConstructionSet(cs)
        sp.setDefaultScheduleSet(ss)

#4 アップロード
print("★ .osm を選択してください")
up = files.upload()
targets = [n for n in up if n.endswith(".osm")]

#5 処理
outs = []
for f in targets:
    m = osd.model.Model.load(osd.path(f)).get()
    assign(m)
    out = f.replace(".osm","_assigned.osm")
    m.save(osd.path(out), True)
    outs.append(out)

#6 ダウンロード
if len(outs)==1:
    files.download(outs[0])
else:
    zipname = f"assigned_{datetime.datetime.now().strftime('%Y%m%d_%H%M%S')}.zip"
    with zipfile.ZipFile(zipname,"w") as z:
        [z.write(o) for o in outs]
    files.download(zipname)
```
